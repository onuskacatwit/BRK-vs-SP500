```python
# David Shneyderman, Individual section:
# Finds the simple moving average over 50 days and the exponential moving average over 200 days for the S&P 500.

#The simple and exponential moving averages are like a trendline but they have a time period over which they are
#averaging the price movements which provides us with a more accurate picture of market momentum given varying timeframes
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as datetime
data = pd.read_csv('Berkshire_vs_500 - Berkshire_vs_500 (16).csv')

data = data.set_index(['date'])
data.index = pd.to_datetime(data.index)

# Simple Moving Average 
def SMA(data, ndays): 
    SMA = pd.Series(data['close_BRK'].rolling(ndays).mean(), name = 'SMA') 
    data = data.join(SMA) 
    return data

# Exponentially-weighted Moving Average 
def EWMA(data, ndays): 
    EMA = pd.Series(data['close_BRK'].ewm(span = ndays, min_periods = ndays - 1).mean(), 
                 name = 'EWMA' ) 
    data = data.join(EMA) 
    return data

close = data['close_BRK']

n = 50
SMA = SMA(data,n)
SMA = SMA.dropna()
SMA = SMA['SMA']

ew = 200
EWMA = EWMA(data,ew)
EWMA = EWMA.dropna()
EWMA = EWMA['EWMA']

plt.figure(figsize=(10,7))

plt.title('Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')

plt.plot(data['close_BRK'],lw=1, label='Close Price')
plt.plot(SMA,'g',lw=1, label='50-day Moving Average')
plt.plot(EWMA,'r', lw=1, label='200-day Exponential Moving Average')

plt.legend()

plt.show()
```
