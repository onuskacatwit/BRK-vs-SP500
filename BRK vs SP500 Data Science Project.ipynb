```python
import pandas as pd
import numpy as np
import datetime as datetime

data = pd.read_csv("Berkshire_vs_500 - Berkshire_vs_500 (11).csv")
data = data.set_index(['date'])


datas = data.columns.values.tolist()

print(datas)
data
```

```python
data.describe()
```

```python
data['percent_change_BRK'] = ((data['next_open_BRK']-data['open_BRK'])/data['open_BRK'])*100
data['percent_change_SP500'] = ((data['next_open_SP']-data['open_SP500'])/data['open_SP500'])*100

data
```

```python
def stringifyTotal(BRK,SP):
     print("Total gain of Berkshire Hathaway: %"+str(BRK)+"\nTotal gain of S&P500: %"+str(SP)+"\nThats a "+ str(abs(BRK-SP)) + "% difference!")
totalBRK = ((data['open_BRK'][len(data)-1]-data['open_BRK'][0])/data['open_BRK'][0])*100
totalSP = ((data['open_SP500'][len(data)-1]-data['open_SP500'][0])/data['open_SP500'][0])*100
stringifyTotal(totalBRK,totalSP)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure(figsize = (20,20))
ax = fig.add_subplot(111)
data.index = pd.to_datetime(data.index)
ax.plot(data.index, data['open_BRK'])
plt.xlabel('Dates (1997-2022)')
plt.ylabel('Opening Prices of BRK')
plt.title('Opening Prices of BRK Over 25 Years')
```

```python
fig = plt.figure(figsize = (20,20))
ax = fig.add_subplot(111)
ax.plot(data.index, data['open_SP500'], color = 'r')
plt.xlabel('Dates (1997-2022)')
plt.ylabel('Opening Prices of S&P500')
plt.title('Opening Prices of S&P500 Over 25 Years')
```

```python
fig = plt.figure(figsize = (20,20))
ax = fig.add_subplot(111)
datax = data[(data.index.year < 2014) & (data.index.year > 2005)]
ax.plot(datax.index, datax['open_SP500'], color = 'r')
plt.xlabel('Dates')
plt.ylabel('Opening Price of S&P500')
```

```python
fig = plt.figure(figsize = (20,20))
ax = fig.add_subplot(111)
ax.plot(datax.index, datax['open_BRK'], color = 'b')
plt.xlabel('Dates')
plt.ylabel('Opening Price of BRK')
```

```python
data_BRK = data.loc[data.groupby(data.index.year)['percent_change_BRK'].idxmax().values]
data_BRK = data_BRK.iloc[:,10:]
data_SP500 = data.loc[data.groupby(data.index.year)['percent_change_SP500'].idxmax().values]
data_SP500 = data_SP500.iloc[:,10:]
data_dates = pd.concat([data_BRK,data_SP500])
data_dates.index = pd.to_datetime(data_dates.index)
data_dates = data_dates.sort_index()
data_dates['later_half'] = data_dates.index.day > 15
data_dates['later_half'].value_counts()
data_dates
```
